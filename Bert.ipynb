```python
from happytransformer import HappyWordPrediction
```

```python
happy_wp = HappyWordPrediction("BERT", "bert-base-uncased")
```

```python
result = happy_wp.predict_mask("I think therefore I [MASK]", top_k=3)
```

```python
print(result)
```

```python
happy_wp.train("train.txt")
```
